```python
# Extraer los datos de natalidad y mortalidad
natalidad = [aqua['Pinar del Ro']['Natalidad'][year] for year in sorted(aqua['Pinar del Ro']['Natalidad'])]
mortalidad = [aqua['Pinar del Ro']['Mortalidad'][year] for year in sorted(aqua['Pinar del Ro']['Mortalidad'])]
aos = sorted(aqua['Pinar del Ro']['Natalidad'].keys())

# Calcular el tamao de las burbujas basado en la diferencia absoluta entre natalidad y mortalidad
tamao_burbujas = [abs(n - m) * 10 for n, m in zip(natalidad, mortalidad)]

# Crear el grfico de dispersin con burbujas
plt.scatter(aos, natalidad, s=tamao_burbujas, alpha=0.5, color='greenyellow', label='Natalidad')
plt.scatter(aos, mortalidad, s=tamao_burbujas, alpha=0.5, color='gray', label='Mortalidad')

# Aadir ttulos y etiquetas
plt.title('Relacin entre Natalidad y Mortalidad en Pinar del Ro (2018-2022)')
plt.xlabel('Aos')
plt.ylabel('Tasa (%)')
plt.legend()
```
