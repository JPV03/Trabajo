```python
#Primero se importan las bibliotecas necesarias
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Luego se cargan los json
with open('aquaman.json', 'r') as hazbinhotel:
    aqua = json.load(hazbinhotel)
    
#Se establecen claves y valores   
provincias = list(aqua['poblacionactualxprovincias'].keys())
poblaciones = list(aqua['poblacionactualxprovincias'].values())

#Para hacer un grafico de barras con las poblaciones provinciales
plt.figure(figsize=(10, 8))
plt.bar(provincias, poblaciones, color='aquamarine')

# Añadir título y etiquetas
plt.title('Población Actual por Provincias en Cuba')
plt.xlabel('Provincias')
plt.ylabel('Población')
plt.xticks(rotation=90)  # Rotar los nombres de las provincias para mejor legibilidad

# Ajustar y mostrar la gráfica
plt.tight_layout()
plt.show()
```

Pinar del Río, ubicada en el extremo occidental de Cuba, se distingue por su rica diversidad cultural y natural. Con una población aproximada de 575 545 habitantes, la provincia se encuentra en un rango medio en cuanto a número de residentes en comparación con otras regiones del país. Este número refleja una comunidad vibrante y diversa, que contribuye significativamente al mosaico demográfico de Cuba.

La Isla de la Juventud, por otro lado, se reconoce como el territorio con la menor cantidad de residentes, lo que le otorga un carácter distintivo dentro del panorama nacional. En contraste, La Habana se erige como la provincia más densamente poblada, un reflejo de su centralidad en la vida económica, social y cultural de la nación, lo que la hacen el destino migratorio interno por excelencia

La provincia de Pinar del Río es conocida por su patrimonio natural, incluyendo el famoso Valle de Viñales, y por su tradición en la producción de tabaco, que es considerada la mejor del mundo. La capital provincial, que lleva el mismo nombre, es un centro de actividad económica y cultural, donde la historia y la modernidad se entrelazan en su arquitectura y en la vida cotidiana de sus habitantes.

```python
municipiosdepinar = aqua['municipiosdepinar']# Extraer los años y los nombres de los municipios
años = list(municipiosdepinar.keys())
municipios = list(municipiosdepinar[años[0]].keys())

# Crear una figura y un conjunto de subplots
fig, axs = plt.subplots(len(años), 1, figsize=(10, len(años)*5))

for i, año in enumerate(años):
    smi_vals = [municipiosdepinar[año][mun]['SMI'] for mun in municipios]
    tmi_vals = [municipiosdepinar[año][mun]['TMI'] for mun in municipios]
    sme_vals = [municipiosdepinar[año][mun]['SME'] for mun in municipios]
    tme_vals = [municipiosdepinar[año][mun]['TME'] for mun in municipios]

    # Configurar las posiciones y anchura de las barras
    width = 0.2
    ind = range(len(municipios))

    # Dibujar las barras para cada tipo de dato
    axs[i].bar(ind, smi_vals, width, label='SMI')
    axs[i].bar([x + width for x in ind], tmi_vals, width, label='TMI')
    axs[i].bar([x + width*2 for x in ind], sme_vals, width, label='SME')
    axs[i].bar([x + width*3 for x in ind], tme_vals, width, label='TME')

    # Configurar los títulos y las etiquetas
    axs[i].set_title('Datos de Migración de ' + año)
    axs[i].set_xticks([x + width*1.5 for x in ind])
    axs[i].set_xticklabels(municipios, rotation=90)
    axs[i].legend()

# Ajustar el layout y mostrar el gráfico
plt.tight_layout()
plt.show()
```

Las causas detrás de este fenómeno migratorio podrían ser tan variadas como los sabores de un buen ajiaco cubano:

1. Económicas: El desempleo o subempleo, como un mal sabor en el paladar, puede empujar a la población a buscar sustento en otras mesas. La inestabilidad económica, como un viento inclemente, puede ahuyentar la inversión y el desarrollo.

2. Sociales: La sed de conocimiento y bienestar puede llevar a las personas a cruzar montañas y mares en busca de educación y servicios de salud. La reunificación familiar, como un imán, puede atraer a individuos hacia aquellos que ya han plantado raíces en nuevos suelos.

3. Políticas: Las políticas migratorias, como las reglas de un juego estricto, pueden limitar la llegada de nuevos rostros. La inestabilidad política, como una tormenta en el horizonte, puede hacer que la población busque refugio lejos de casa.

4. Ambientales: Los desastres naturales, como el huracán Ian en 2022, pueden convertir hogares en recuerdos, forzando a la gente a reconstruir sus vidas en otros lugares.

En la encuesta realizada, un 76.4% de los encuestados manifestó deseos de emigrar, y de estos, un 95.25% aspiraba a cruzar fronteras internacionales, dejando solo a dos valientes dispuestos a cambiar de aires dentro de la isla, uno por la llamada de la academia y otro por la sirena de la economía.

Entonces, ¿cuáles son las consecuencias de este vaivén migratorio en la provincia de Pinar del Río?

```python
# Extraer los datos de natalidad y mortalidad
natalidad = [aqua['Pinar del Río']['Natalidad'][year] for year in sorted(aqua['Pinar del Río']['Natalidad'])]
mortalidad = [aqua['Pinar del Río']['Mortalidad'][year] for year in sorted(aqua['Pinar del Río']['Mortalidad'])]
años = sorted(aqua['Pinar del Río']['Natalidad'].keys())

# Calcular el tamaño de las burbujas basado en la diferencia absoluta entre natalidad y mortalidad
tamaño_burbujas = [abs(n - m) * 10 for n, m in zip(natalidad, mortalidad)]

# Crear el gráfico de dispersión con burbujas
plt.scatter(años, natalidad, s=tamaño_burbujas, alpha=0.5, color='greenyellow', label='Natalidad')
plt.scatter(años, mortalidad, s=tamaño_burbujas, alpha=0.5, color='gray', label='Mortalidad')

# Añadir títulos y etiquetas
plt.title('Relación entre Natalidad y Mortalidad en Pinar del Río (2018-2022)')
plt.xlabel('Años')
plt.ylabel('Tasa (%)')
plt.legend()
```


En los últimos 5 años en Pinar del Río, se han producido una serie de cambios demográficos, incluyendo la disminución de las tasas de natalidad y el aumento de las tasas de mortalidad.

La disminución de las tasas de natalidad se debe a varios factores, como el aumento de la edad materna, el acceso a la planificación familiar y los cambios en las normas sociales. Por otro lado, el aumento de las tasas de mortalidad se relaciona en parte con el envejecimiento de la población y la mayor prevalencia de enfermedades crónicas. Es importante destacar que la pandemia también ha influido en este último aspecto. Por ejemplo, la tercera hija de Fernando, quien era doctora, falleció trágicamente durante la pandemia, dejando a sus dos hijos pequeños al cuidado de sus abuelos.
