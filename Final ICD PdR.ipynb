# **Entre Hojas de Tabaco y Sueños Errantes: Crónicas de Migración en Pinar del Río**

*"La migración es el espejo de los tiempos, reflejando la búsqueda eterna del ser humano por un lugar al que llamar hogar"*

 En el corazón de Viñales, donde emergen los mogotes como guardianes eternos de piedra caliza, se alza la morada de Fernando. Este lugar, abrazado por campos de tabaco y tierra fértil de tonalidades rojizas, resguarda los recuerdos vivientes de la historia y la identidad cultural de la región.

El primogénito de Fernando, arraigado en La Florida desde hace quince años, representa la esencia viva de la diáspora cubana. El segundo vástago, que partió hacia el municipio de Pinar del Río tras su enlace matrimonial, ha sido impregnado con el espíritu de su nuevo hogar. La tercera hija, cuya presencia se desvaneció prematuramente, dejó un legado de amor y valentía en sus dos pequeños hijos, quienes ahora crecen bajo el cuidado amoroso de sus abuelos. Y la más joven, Yaime, tras mudarse a La Habana para estudiar Arquitectura y decidir quedarse, encarna los sueños y aspiraciones de una nueva generación.

¿Podrá la familia de Fernando servir de guía para desentrañar las características de la migración en Pinar del Río? ¿Acaso sus relatos revelan cómo se entrelazan los lazos familiares, la cultura y la tierra en el bordado migratorio?


```python
#Primero se importan las bibliotecas necesarias
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Luego se cargan los json
with open('aquaman.json', 'r') as hazbinhotel:
    aqua = json.load(hazbinhotel)
    
#Se establecen claves y valores   
provincias = list(aqua['poblacionactualxprovincias'].keys())
poblaciones = list(aqua['poblacionactualxprovincias'].values())

#Para hacer un grafico de barras con las poblaciones provinciales
plt.figure(figsize=(10, 8))
plt.bar(provincias, poblaciones, color='skyblue')

# Añadir título y etiquetas
plt.title('Población Actual por Provincias en Cuba')
plt.xlabel('Provincias')
plt.ylabel('Población')
plt.xticks(rotation=90)  # Rotar los nombres de las provincias para mejor legibilidad

# Ajustar y mostrar la gráfica
plt.tight_layout()
plt.show()
```

Al examinar la población de las provincias de Cuba, notamos que Pinar del Río se sitúa en un rango medio en cuanto a habitantes, mientras que la Isla de la Juventud, como municipio especial, se distingue por ser el lugar con la menor cantidad de residentes. Contrastando con esto, La Habana emerge como la provincia más densamente poblada, lo que refleja su importancia en la esfera económica, social y cultural de la nación.

La capital, La Habana, ejerce una fuerte atracción, impulsada por factores como el empleo, la educación avanzada y los servicios de salud de calidad, según se reporta en el último anuario estadístico. Estos elementos la convierten en un destino preferente para aquellos que buscan mejorar su calidad de vida. Además, su vibrante actividad turística y su posición como centro de desarrollo tecnológico y de innovación fomentan un flujo constante de migración interna hacia la ciudad.

Este patrón migratorio se ve reflejado en la decisión de Yaime, la hija menor de Fernando, quien se trasladó a La Habana para perseguir sus aspiraciones académicas y profesionales. 

```python
datos_emigracion = aqua['MMIemigracion']
provincias = datos_emigracion.keys()

# Crear el gráfico
plt.figure(figsize=(10, 5))

# Generar un color diferente para cada provincia
colores = plt.get_cmap('hsv', len(provincias))

# Iterar sobre cada provincia y plotear los datos
for i, provincia in enumerate(provincias):
    anos = [dato['ano'] for dato in datos_emigracion[provincia]]
    cantidades = [dato['cantidad'] for dato in datos_emigracion[provincia]]
    plt.scatter(anos, cantidades, color=colores(i), label=provincia)

# Añadir título y etiquetas
plt.title('Emigración desde Pinar del Río a otras provincias')
plt.xlabel('Año')
plt.ylabel('Cantidad de emigrantes')
plt.legend(loc='upper left',
           bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

En 2014, se registró un descenso notable en la emigración de pinareños a otras provincias, una tendencia que solo se replicaría durante el aislamiento de la pandemia en 2020. Sin embargo, en 2015, la migración hacia La Habana alcanzó su punto máximo, con 1380 personas trasladándose, coincidiendo con el inicio de los estudios de arquitectura de Yaime en la capital. Asi mismo se puede notar como la cifra de emigrantes hacia La Habana siempre suele ser mayor a 1000. Este pico refleja la importancia de las oportunidades educativas en los patrones migratorios. Por otro lado, la migración hacia Artemisa también es significativa, en parte debido a la división política-administrativa que anteriormente integraba parte de Artemisa a Pinar del Río, manteniendo fuertes lazos comunitarios y familiares que facilitan la movilidad entre estas provincias.

```python
relacion = aqua['MMIPinan']['Emigracion']

# Preparar los datos para los gráficos
anos = [dato['ano'] for dato in relacion]
hombres = [dato['hombres'] for dato in relacion]
mujeres = [dato['mujeres'] for dato in relacion]

# Crear el gráfico de barras apiladas para emigración
plt.figure(figsize=(10, 5))
plt.bar(anos, hombres, color='blue', label='Hombres')  # Azul para hombres
plt.bar(anos, mujeres, bottom=hombres, color='pink', label='Mujeres')  # Rosado para mujeres

# Añadir título y etiquetas
plt.title('Emigración de Pinar del Río por género y año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.legend()

# Mostrar el gráfico
plt.show()
```

La migración desde Pinar del Río hacia otras provincias cubanas presenta patrones interesantes cuando se analiza por género y año. Aunque la diferencia no es abrumadora, se observa que las mujeres tienden a emigrar más que los hombres, salvo algunas excepciones como los años 2015 y 2017. Este fenómeno puede estar influenciado por diversos factores, incluyendo la búsqueda de oportunidades laborales, educativas o la reunificación familiar.

Estos datos podrian sugerir que las mujeres jóvenes son particularmente proclives a buscar nuevos horizontes, como lo hizo la hija de nuestro querido Fernando, lo que podría estar relacionado con el deseo de mejorar su situación socioeconómica o de seguir estudios superiores en otras provincias.

Es importante considerar cómo las normas sociales y las expectativas de género pueden influir en la decisión de migrar. Además, la inclusión de consideraciones de género en la planificación de políticas puede contribuir al empoderamiento social y económico de las mujeres y promover la igualdad de género.

```python
datos_inmigracion = aqua['inmigracion']
provincias = datos_inmigracion.keys()

# Crear el gráfico
plt.figure(figsize=(10, 5))

# Generar un color diferente para cada provincia
colores = plt.get_cmap('hsv', len(provincias))

# Iterar sobre cada provincia y plotear los datos
for i, provincia in enumerate(provincias):
    anos = [dato['ano'] for dato in datos_inmigracion[provincia]]
    cantidades = [dato['cantidad'] for dato in datos_inmigracion[provincia]]
    plt.scatter(anos, cantidades, color=colores(i), label=provincia)

# Añadir título y etiquetas
plt.title('Inmigracion a Pinar del Rio')
plt.xlabel('Año')
plt.ylabel('Cantidad de inmigrantes')
plt.legend(loc='upper left',
           bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```

La inmigración hacia Pinar del Río, aunque menor en comparación con la emigración, es un componente importante en la dinámica poblacional de la provincia. Pinar del Río tiene un saldo migratorio negativo, similar al de las provincias de la región oriental, a diferencia de las tendencias de Occidente. Durante la pandemia, hubo una disminución notable en la inmigración, lo que refleja las restricciones de movimiento y el impacto económico global. Las principales fuentes de inmigración hacia Pinar del Río son La Habana y Artemisa, que destaca la relación continua entre estas provincias, en especial con esta ultima, con una media de 420 inmigrantes provenientes de esta provincia entre los años 2011 y 2022

En 2014, se observó un incremento en la inmigración desde Mayabeque, lo que representa una excepción en comparación con los lazos migratorios generalmente más débiles que Pinar del Río tiene con el resto de las provincias

```python
relacion = aqua['MMIPinan']['inmigracion']

# Preparar los datos para los gráficos
anos = [dato['ano'] for dato in relacion]
hombres = [dato['hombres'] for dato in relacion]
mujeres = [dato['mujeres'] for dato in relacion]

# Crear el gráfico de barras apiladas para la inmigración
plt.figure(figsize=(10, 5))
plt.bar(anos, hombres, color='blue', label='Hombres')  # Azul para hombres
plt.bar(anos, mujeres, bottom=hombres, color='pink', label='Mujeres')  # Rosado para mujeres

# Añadir título y etiquetas
plt.title('Inmigracion a Pinar del Río por género y año')
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.legend()

# Mostrar el gráfico
plt.show()
```

La inmigración hacia Pinar del Río desde otras provincias se caracteriza por una distribución equitativa entre géneros, reflejando un equilibrio en la movilidad poblacional. Esta paridad migratoria es un testimonio de la igualdad de oportunidades y aspiraciones que impulsan tanto a hombres como a mujeres a buscar nuevos horizontes dentro de la isla. Tal simetría en los flujos migratorios internos es un reflejo de la armonía social y la uniformidad de motivaciones que trascienden las diferencias de género, delineando un panorama de migración donde la equidad es la protagonista.

```python
saldomigratorio = aqua['Cubasyt']['saldomigratorio']
tasamigratoria = aqua['Cubasyt']['tasamigratoria']

# Crear un DataFrame de pandas con los datos
df = pd.DataFrame({
    'Año': saldomigratorio.keys(),
    'Saldo Migratorio': saldomigratorio.values(),
    'Tasa Migratoria': tasamigratoria.values()
})

# Ordenar el DataFrame por año
df.sort_values('Año', inplace=True)


# Mostrar la tabla
print(df)
```

La tabla de saldo y tasa migratoria de Cuba muestra una tendencia general hacia un saldo negativo, indicando una mayor emigración que inmigración. Sin embargo, en 2013 y 2014, la tendencia se invierte con un saldo migratorio positivo. Este cambio puede estar vinculado a las políticas de flexibilización migratoria implementadas por el gobierno cubano en 2013, que permitieron a los cubanos vivir en el exterior hasta dos años sin perder su residencia y facilitaron los procesos de repatriación. Estas medidas podrían haber incentivado a muchos cubanos, incluido José Carlos, a considerar el retorno a Cuba o a posponer sus planes de emigración.

La historia de José Carlos, el hijo mayor de Fernando, refleja estas dinámicas migratorias. Aunque contribuyó al saldo migratorio negativo de Cuba al emigrar a Florida, las políticas migratorias más abiertas y las mejoras económicas o cambios en las relaciones internacionales durante 2013 y 2014 podrían haber influido en su decisión de regresar a Cuba o en la de otros cubanos en situaciones similares.

```python
datos_pinar = aqua['PinardelRio']

# Crear un DataFrame para cada conjunto de datos
df_saldo_total = pd.DataFrame(list(datos_pinar['saldomigratoriot'].items()), columns=['Año', 'Saldo Total'])
df_saldo_interno = pd.DataFrame(list(datos_pinar['saldomigratorioi'].items()), columns=['Año', 'Saldo Interno'])
df_saldo_externo = pd.DataFrame(list(datos_pinar['saldomigratorioe'].items()), columns=['Año', 'Saldo Externo'])
df_tasa_total = pd.DataFrame(list(datos_pinar['tasamigratoriat'].items()), columns=['Año', 'Tasa Total'])
df_tasa_interna = pd.DataFrame(list(datos_pinar['tasamigratoriai'].items()), columns=['Año', 'Tasa Interna'])
df_tasa_externa = pd.DataFrame(list(datos_pinar['tasamigratoriae'].items()), columns=['Año', 'Tasa Externa'])

# Combinar todos los DataFrames en uno solo para una vista completa
df_combinado = pd.concat([df_saldo_total.set_index('Año'), df_saldo_interno.set_index('Año'), df_saldo_externo.set_index('Año'),
                          df_tasa_total.set_index('Año'), df_tasa_interna.set_index('Año'), df_tasa_externa.set_index('Año')], axis=1)

# Mostrar la tabla combinada
print(df_combinado)
```

El análisis de los datos de migración de Pinar del Rio revela varias tendencias y patrones interesantes:

1. Saldo Migratorio Total: Ha habido una tendencia general a la disminución del saldo migratorio total desde 2011 hasta 2021, con una leve mejora en 2013 y 2014. El año 2012 registró el saldo más negativo con -3769, mientras que 2014 tuvo el menor con -376.

2. Saldo Migratorio Interno y Externo: Los saldos internos y externos contribuyen al saldo total negativo, pero en 2013 y 2014, el saldo externo fue positivo, lo que indica un mayor número de inmigrantes que de emigrantes en esos años.

3. Tasa Migratoria Total: La tasa migratoria total ha fluctuado a lo largo de los años, alcanzando su punto más bajo en 2012 con -6.4 y mejorando en los años siguientes. Sin embargo, sigue siendo negativa, lo que sugiere que la población de Pinar del Rio ha experimentado una disminución neta debido a la migración.

4. Tasa Migratoria Interna y Externa: La tasa migratoria interna ha sido consistentemente negativa, lo que indica una pérdida de población dentro del país. La tasa migratoria externa muestra una variabilidad significativa, con años de tasas positivas y negativas.
 
 Las tendencias migratorias negativas en Pinar del Rio, como en cualquier región, pueden ser el resultado de una combinación de factores socioeconómicos, políticos y ambientales. Aquí hay algunas posibles causas que podrían explicar los datos observados:

1. Económicas:
   - Desempleo o subempleo: La falta de oportunidades laborales o trabajos mal remunerados pueden impulsar a la población a buscar mejores condiciones en otros lugares.
   - Inestabilidad económica: La inflación, la deuda y otros problemas económicos pueden disuadir la inversión y el desarrollo, afectando negativamente la migración.
2. Sociales:
   - Educación y servicios: La búsqueda de mejor educación o servicios de salud puede motivar a las personas a migrar.
   - Reunificación familiar: Las personas pueden emigrar para reunirse con familiares que ya se han mudado a otras áreas o países.
3. Políticas:
   - Políticas migratorias restrictivas: Las leyes que limitan la inmigración pueden disminuir el saldo migratorio positivo.
   - Conflictos y estabilidad: La inestabilidad política o los conflictos pueden forzar a la población a buscar seguridad en otros lugares.
4. Ambientales:
   - Desastres naturales: Los eventos climáticos extremos o la degradación ambiental pueden hacer que las áreas sean menos habitables., como el caso del huracan Ian en el año 2022
